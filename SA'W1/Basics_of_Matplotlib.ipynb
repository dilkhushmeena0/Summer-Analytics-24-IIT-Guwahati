```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
plt.plot([1,3,2,4])
plt.show()
```

```python
plt.plot([1,2,3,4],[1,4,9,16])
plt.title('Test plot',fontsize=16,color='r')
plt.xlabel('number n')
plt.ylabel('n^2')
plt.show()
```

```python
plt.figure(figsize=(2,5))
plt.plot([1,2,3,4],[1,4,9,16],'ro')
plt.axis([0,6,0,20])
plt.annotate('square it',(3,6))
plt.show()
```

```python
plt.clf()
x=np.arange(4)
y=[8.8,5.6,4.4,7.8]
plt.xticks(x,('dfs','sfs','sgs','wwf'))
plt.bar(x,y)
# plt.bar(x,y,color=['lime','g','r','b'])
plt.show()
```

```python
d = {'Red O':np.random.rand(20),
     'Green O':np.random.rand(20)}
df=pd.DataFrame(d)
df.plot(style=['ro','go'])
plt.show()
```

```python
ts = pd.Series(np.random.randn(180),index=pd.date_range('1/1/2018',periods=180))
df = pd.DataFrame(np.random.randn(180,4),index=ts.index,columns=list('ABCD'))
df.cumsum().plot()
plt.show()
```

```python
N=50
x=np.random.rand(N)
y=np.random.rand(N)
colors=np.random.rand(N)
sizes=(30*np.random.rand(N))**2 #0 to 14 radii
plt.scatter(x,y,s=sizes,c=colors,alpha=0.5) # alpha is the transparency
plt.show()
```

```python
# we can show multiple plots in single plot
```

```python
fig = plt.figure()
fig.suptitle('Subplots',fontsize=24)
fig.add_subplot(221)
plt.plot([np.log(n) for n in range(1,10)])
fig.add_subplot(222, facecolor='r')
fig.add_subplot(223)
fig.add_subplot(224)
plt.show()
```

```python
fig,plots = plt.subplots(2,sharex=True)
fig.suptitle('Sharing X axis',fontsize=16)
x = range(0,200,5)
y = [n**0.8 for n in x]
plots[0].plot(x,y,color='r')
plots[1].scatter(x,y)
plt.show()
```

```python
plt.figure(figsize=(4,3),dpi=480)
plt.plot([245,170,253,243,145,156,244])
plt.axis([0,7,0,300])
plt.title('Flight Data',fontsize=16)
plt.xlabel('Speed')
plt.savefig('Flights.png')
plt.show()
```
