```python
import matplotlib.pyplot as plt
fig=plt.figure(figsize=(15,5))
x=np.arange(0,3*np.pi,0.1)
y_sin=np.sin(x)
y_cos=np.cos(x)
y_tan=np.tan(x)

plt.subplot(1,3,1)
plt.plot(x,y_sin)
plt.title('sine')

plt.subplot(1,3,3)
plt.plot(x,y_cos)
plt.title('cos')

plt.subplot(1,3,2)
plt.plot(x,y_tan)
plt.title('tan')
```
