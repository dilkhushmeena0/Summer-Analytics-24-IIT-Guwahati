```python
#Standardization
import pandas as pd
df = pd.read_csv('C:titanic_train.csv',usecols=['Pclass','Age','Fare','Survived'])
df.head()
```

```python
df['Age'].fillna(df.Age.median(),inplace=True)
df.isnull().sum()
```

```python
x=df.iloc[:,1:]
y=df.iloc[:,0]
x
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test =  train_test_split(x,y,test_size=0.33,random_state=42)
x_train
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_train_scaled
```

```python
x_test_scaled = scaler.transform(x_test)
x_test_scaled
```

```python
# Min Max Scaling-- Used in Deep Learning Technique
from sklearn.preprocessing import MinMaxScaler
min_max = MinMaxScaler()
df_minmx = pd.DataFrame(min_max.fit_transform(x_train))
df_minmx.head()
```

```python
# Robust Scaler
from sklearn.preprocessing import RobustScaler
scaler=RobustScaler()
df_robust_scaler = pd.DataFrame(scaler.fit_transform(x_train))
df_robust_scaler
```
